```python
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

```python
df_wire1= pd.read_csv('./BiotSavartLaw_wire1.csv')
df_wire2= pd.read_csv('./BiotSavartLaw_wire2.csv')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')
ax.plot(df_wire1.iloc[:,0], df_wire1.iloc[:,1], df_wire1.iloc[:,2])
ax.plot(df_wire2.iloc[:,0], df_wire2.iloc[:,1], df_wire2.iloc[:,2])
```
