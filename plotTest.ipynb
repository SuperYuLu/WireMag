```python
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

```python
df_wire= pd.read_csv('./wire.csv')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')
ax.plot(df_wire.iloc[:,0], df_wire.iloc[:,1], df_wire.iloc[:,2])
```

```python
df_mesh = pd.read_csv('./meshgrid.csv')
```

```python
df_field = pd.read_csv('x_y_z_Babs.csv')
```
